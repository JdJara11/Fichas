```python
def truth_table_conjunction():
    print("p | q | p ‚àß q (AND)")
    for p in [True, False]:
        for q in [True, False]:
            print(f"{p} | {q} | {p and q}")

def truth_table_disjunction():
    print("p | q | p ‚à® q (OR)")
    for p in [True, False]:
        for q in [True, False]:
            print(f"{p} | {q} | {p or q}")

def truth_table_implication():
    print("p | q | p ‚Üí q (not OR)")
    for p in [True, False]:
        for q in [True, False]:
            print(f"{p} | {q} | {not p or q}")

def truth_table_biconditional():
    print("p | q | p ‚Üî q (EQUAL)")
    for p in [True, False]:
        for q in [True, False]:
            print(f"{p} | {q} | {p == q}")

def truth_table_xor():
    print("p | q | p ‚äï q (NOT EQUAL)")
    for p in [True, False]:
        for q in [True, False]:
            print(f"{p} | {q} | {p != q}")

def truth_table_xand():
    print("p | q | p XAND q")
    print("-" * 15)
    for p in [True, False]:
        for q in [True, False]:
            a = p != q
            anot = not a
            print(f"{p} | {q} | {anot}")  # XAND es NOT XNOR
# Llamar las funciones para ver las tablas
truth_table_conjunction()
truth_table_disjunction()
truth_table_implication()
truth_table_biconditional()
truth_table_xor()
truth_table_xand()
```
