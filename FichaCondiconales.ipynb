# CONDICIONALES if/elif/else

1. **Describa el por qué y para qué se utiliza.**  


 En python las estructuras condicionales permiten ejecutar diferentes bloques de código según si una condición es verdadera o falsa. Se usan para evitar la repetición de líneas de código.

2. **Utiliza el código explicado en clase y agregue el/los ejemplos**

```python
seleccion = input("Ingrese la operacion a realizar: ")
a = int(input("Ingrese un numero: "))
b = int(input("Ingrese otro numero: "))


if "resta" == seleccion:
    print("La resta es: ", (a-b))
elif "multiplicacion" == seleccion:
    print("La multiplicacion es: ", (a*b))
elif "division" == seleccion:
    print("La division es: ", (a/b))
elif "suma" == seleccion:
    print("La suma es: ", (a+b))
else:
    print("Operacion no valida")
```

```python
numero = int(input("Ingresa un número: "))

# Evaluar si es positivo, negativo o cero
if numero > 0:
    print("El número es positivo.")
elif numero < 0:
    print("El número es negativo.")
else:
    print("El número es cero.")
```
