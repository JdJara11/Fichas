# CICLO for

1. **Describa el por qué y para qué se utiliza.**  


El ciclo for se usa para iterar sobre una secuencia de elementos como listas o rangos. Evita repetición de código y facilita las tareas repetitivas.

2. **Utiliza el código explicado en clase y agregue el/los ejemplos**

```python
calificaciones = {
    "Ana": 85,
    "Luis": 55,
    "Carlos": 90,
    "Maria": 45,
    "José": 70
}

# Recorremos el diccionario con un for
for estudiante, calificacion in calificaciones.items():
    if calificacion >= 60:
        print(f"{estudiante} aprobó con una calificación de {calificacion}")
    else:
        print(f"{estudiante} reprobó con una calificación de {calificacion}")
```

```python
estudiantes = {
    "Ana": {"Matemáticas", "Ciencias", "Historia"},
    "Luis": {"Ciencias", "Física"},
    "Carlos": {"Matemáticas", "Física", "Química"},
    "Maria": {"Arte"},
    "José": {"Historia", "Arte"}
}

# Recorremos el diccionario con un for
for estudiante, cursos in estudiantes.items():
    if len(cursos) > 2:
        print(f"{estudiante} está tomando más de dos cursos: {cursos}")
```
