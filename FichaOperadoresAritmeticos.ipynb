```python
a = 10
b = 3

suma = a + b
resta = a - b
multiplicacion = a * b
division = a / b  # División flotante
division_entera = a // b  # División entera
modulo = a % b  # Resto de la división
potencia = a ** b  # Exponente

print(f"Suma: {suma}, Resta: {resta}, Multiplicación: {multiplicacion}")
print(f"División: {division}, División Entera: {division_entera}")
print(f"Módulo: {modulo}, Potencia: {potencia}")
```
