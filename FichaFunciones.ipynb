1. **Describa el por qué y para qué se utiliza.**  


Las funciones permiten reutilizar código y organizar mejor los programas. Pueden recibir parámetros y devolver valores. Sirve para definir un proceso de entrada, proceso y salida. Al declarar la función se establece un scope, el cual permite proteger el contenido real de las variables internamente, sin que se afecte lo esterno. Las funciones puede requerir uno o muchos parametros y estos deben afectar el scope interno de la función.
