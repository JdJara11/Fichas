```python
# Declaración de variables
nombre = "Ana"
edad = 25
altura = 1.65

# Impresión de variables
print("Nombre:", nombre)
print("Edad:", edad)
print("Altura:", altura)

# Tipos de datos
print(type(nombre))  # str
print(type(edad))    # int
print(type(altura))  # float
```
